# About the dataset
Customer satisfaction scores from 120,000+ airline passengers, including additional information about each passenger, their flight, and type of travel, as well as ther evaluation of different factors like cleanliness, comfort, service, and overall experience</h1><br>
1.ID: Unique passenger identifier<br>
2.Gender: Gender of the passenger (Female/Male)<br>
3.Age: Age of the passenger<br>
4.Customer Type: Type of airline customer (First-time/Returning)<br>
5.Type of Travel: Purpose of the flight (Business/Personal)<br>
6.Class: Travel class in the airplane for the passenger seat<br>
7.Flight Distance: Flight distance in miles<br>
8.Departure Delay: Flight departure delay in minutes<br>
9.Arrival Delay: Flight arrival delay in minutes<br>
10.Departure and Arrival Time Convenience: Satisfaction level with the convenience of the flight departure and arrival times from 1 (lowest) to 5 (highest) - 0 means "not applicable"<br>
11.Ease of Online Booking: Satisfaction level with the online booking experience from 1 (lowest) to 5 (highest) - 0 means "not applicable"<bt>
12.Check-in Service: Satisfaction level with the check-in service from 1 (lowest) to 5 (highest) - 0 means "not applicable"<br>
13.Online Boarding: Satisfaction level with the online boarding experience from 1 (lowest) to 5 (highest) - 0 means "not applicable"<br>
14.Gate Location: Satisfaction level with the gate location in the airport from 1 (lowest) to 5 (highest) - 0 means "not applicable"<br>
15.On-board Service: Satisfaction level with the on-boarding service in the airport from 1 (lowest) to 5 (highest) - 0 means "not applicable"<br>
16.Seat Comfort: Satisfaction level with the comfort of the airplane seat from 1 (lowest) to 5 (highest) - 0 means "not applicable"<br>
17.Leg Room Service: Satisfaction level with the leg room of the airplane seat from 1 (lowest) to 5 (highest) - 0 means "not applicable"<br>
18.Cleanliness: Satisfaction level with the cleanliness of the airplane from 1 (lowest) to 5 (highest) - 0 means "not applicable"<br>
19.Food and Drink: Satisfaction level with the food and drinks on the airplane from 1 (lowest) to 5 (highest) - 0 means "not applicable"<br>
20.In-flight Service: Satisfaction level with the in-flight service from 1 (lowest) to 5 (highest) - 0 means "not applicable"<br>
21.In-flight Wifi Service: Satisfaction level with the in-flight Wifi service from 1 (lowest) to 5 (highest) - 0 means "not applicable"<br>
22.In-flight Entertainment: Satisfaction level with the in-flight entertainment from 1 (lowest) to 5 (highest) - 0 means "not applicable"<br>
23.Baggage Handling: Satisfaction level with the baggage handling from the airline from 1 (lowest) to 5 (highest) - 0 means "not applicable"<br>
24.Satisfaction: Overall satisfaction level with the airline (Satisfied/Neutral or unsatisfied)<br>

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler(copy=True, with_mean=True, with_std=True)
x = scaler.fit_transform(x)
```

```python

def model(m):
    score1=cross_validate (m,x,y,cv=10,return_train_score=True)
    score2=cross_validate (m,x,y,cv=10, scoring=('r2', 'neg_mean_squared_error'),return_train_score=True)
    check=pd.DataFrame({'test_score':score1['test_score'],'train_score':score1['train_score'],'fit_time':score1['fit_time'],'train MSE value ':score2['train_neg_mean_squared_error'],
                        'test MSE value ':score2['test_neg_mean_squared_error'],'train R2 values': score2['train_r2'],'test R2 values':score2['test_r2']})
    print(check)
    
```

```python
from sklearn import  metrics
results = pd.DataFrame(columns=['Model Name','Model state','Train-set Score','Test-set Score', 'Satisfied Accuracy', 'Dissatisfied Accuracy'])
def fit_model_result(x_train, y_train, x_test, y_test, model, model_name, model_change):
    reg = model
    reg.fit(x_train,y_train)
    train_score = reg.score(x_train,y_train)*100
    test_score = reg.score(x_test,y_test)*100
    y_pred = reg.predict(x_test)
    k = metrics.confusion_matrix(y_test, y_pred,normalize='true')*100
    y_pred = y_pred == y_test
    # random indexing for plotting
    indx = np.random.rand(len(y_pred))
    data = {'Actual':y_test,'Predict':y_pred,'index':indx}
    comp = pd.DataFrame(data)
    results.loc[len(results.index)] = [model_name, model_change, train_score, test_score,round(k[0][0],2),round(k[1][1],2)]
    print(results.iloc[-1,:])
    fig, ax = plt.subplots(figsize=(10, 8))
    metrics.plot_confusion_matrix(reg, x_test, y_test, cmap='Greens', normalize='true',ax=ax)
    plt.show()
```

```python
x_train , x_test , y_train , y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
fit_model_result(x_train, y_train, x_test, y_test, LogisticRegression(), 'Logistic Regression', 'Original')
```

```python
fit_model_result(x_train, y_train, x_test, y_test, KNeighborsClassifier(n_neighbors=7), 'KNN Classifier', 'Original')
```

```python
fit_model_result(x_train, y_train, x_test, y_test, DecisionTreeClassifier(max_features=16, max_depth=12, random_state=42), 'DT Classifier', 'Original')
```

```python
fit_model_result(x_train, y_train, x_test, y_test, RandomForestClassifier(max_features=16, max_depth=12, random_state=42), 'RF Classifier', 'Original')
```
